```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image
image_path = "/content/Copy of stage_1_4.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was read successfully
if image is None:
    print("Image not found!")
else:
    # Define the kernel size for morphological operations
    kernel_size = (5, 5)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernel_size)

    # Apply morphological operations
    opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    erosion = cv2.erode(image, kernel)
    dilation = cv2.dilate(image, kernel)

    # Display the original image and processed images with titles
    cv2_imshow(image)
    cv2_imshow(opening)
    cv2_imshow(closing)
    cv2_imshow(erosion)
    cv2_imshow(dilation)


```

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image in color
image_path = "/content/Copy of stage_1_4.jpg"
image = cv2.imread(image_path)

# Check if the image was read successfully
if image is None:
    print("Image not found!")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the image to create a binary image
    _, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

    # Find connected components
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_image, connectivity=8)

    # Display the number of connected components
    print(f"Number of connected components: {num_labels}")

    # Draw rectangles around connected components
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the image with rectangles around connected components
    cv2_imshow(image)
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image in color
image_path = "/content/Copy of stage_1_4.jpg"
image = cv2.imread(image_path)

# Check if the image was read successfully
if image is None:
    print("Image not found!")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the image to create a binary image
    _, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

    # Find connected components
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_image, connectivity=8)

    # Display the number of connected components
    print(f"Number of connected components: {num_labels}")

    # Initialize a list to store major connected components
    major_components = []

    # Define a threshold for the minimum area of connected components
    min_area_threshold = 350  # Adjust this threshold as needed

    # Find major connected components based on area
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        if area > min_area_threshold:
            major_components.append((x, y, w, h))

    # Draw rectangles around major connected components
    for component in major_components:
        x, y, w, h = component
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the image with rectangles around major connected components
    cv2_imshow(image)
```

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Initialize the corner detection algorithm (e.g., Harris corner detection)
corner_detector = cv2.cornerHarris(gray_image, 2, 3, 0.04)  # Adjust parameters as needed

# Threshold the corner response to find strong corners
threshold = 0.01 * corner_detector.max()
corner_image = image.copy()
corner_image[corner_detector > threshold] = [0, 0, 255]  # Mark corners in red

# Display the result
cv2_imshow(corner_image)
```
